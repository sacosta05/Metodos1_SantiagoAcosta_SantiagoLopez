```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,400)
h = 0.05

def derivada_central(f, x, h):
    return (f(x+h) - f(x-h)) / (2*h)

def f(x):
    return 1/np.sqrt(1+np.exp(-x**2))

def derivada_real(x):
    return (x*np.exp(-x**2))/(1+np.exp(-x**2))**(3/2)
```

```python
plt.plot(x,f(x), c='darkblue', linewidth=2)
plt.title("Funci√≥n f(x)", size=18, c='darkblue')
plt.show()
```

```python
plt.plot(x,derivada_real(x),c='darkred', linewidth=2)
plt.title("Funci√≥n f '(x)", size=18, c='darkred')
plt.show()
```

```python
plt.plot(x,np.abs(derivada_real(x)-derivada_central(f,x,h)), c='darkgreen', linewidth=2)
plt.title("Estimaci√≥n del error en cada punto nodal.", size=18, c='darkgreen')
plt.show()
```
