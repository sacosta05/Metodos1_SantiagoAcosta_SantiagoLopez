**Números primos**

1. Escriba un código que calcule los primeros 1000 números primos.

2. Imprima en pantalla los primos 10 números primos.

3. Graficar los números como funcio ́n de su posición, es decir, 2 es el primero, 3 es el segundo,...

```python
import numpy as np
import matplotlib.pyplot as plt

## Codigo para encontrar los 1000 primeros números primos

def num_primos(n):
    if (n % 2 == 0):
        return False
    for i in range(3, int(n**0.5 + 1), 2):
        if (n % i == 0):
            return False
    return True

def imprimir_primos(n):
    primos = [2]
    num = 3
    while (len(primos) < n):
        if(num_primos(num)):
            primos.append(num)
        num = num + 2
    return primos

print(imprimir_primos(1000))

```

```python
## Utilizando el mismo código anterior podmeos encontrar los 10 primeros números primos de los enteros.

print(imprimir_primos(10))
```

```python
x= list(range(1000))
y=(imprimir_primos(1000))

plt.plot(x , y, c='g', label='Grafica de los 1000 primeros números primos')
plt.legend()
plt.show()
```
