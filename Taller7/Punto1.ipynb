```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from tqdm import tqdm
```

```python
x = np.linspace(0,10,100)
y1 = 2*x-2
y2 = (1-x)/2
y3 = 4-x
```

```python
A = np.array([[2,-1],[1,2],[1,1]])
AT = np.dot(A.T,A)
b = np.array([[2],[1],[4]])
bT = np.dot(A.T,b)

soluc = np.linalg.solve(AT,bT)
dist = np.dot(A,soluc)-b
dist_ = np.linalg.norm(dist)
print("Punto solución:",soluc[0],",",soluc[1])
print("Distancia al punto:",dist_)
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(1, 1, 1)
plt.plot(x,y1,c = 'b')
plt.plot(x,y2,c = 'orange' )
plt.plot(x,y3, c= 'r')
plt.scatter(soluc[0],soluc[1],c='c')
plt.show()
```

```python
x = np.arange(-5, 5, 0.03)
y = np.arange(-5, 5, 0.03)
X,Y = np.meshgrid(x, y)
distanc = np.zeros((np.shape(x)[0], np.shape(y)[0]))

mind = 0

for i in range(x.size):
    for j in range(y.size):
        vect = np.array([[x[i]],[y[j]]])
        D = np.dot(A,vect)-b
        d = np.linalg.norm(D)
        distanc[i,j] = d
```

```python
fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(projection='3d')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("d(x*)")
sup = ax.plot_surface(X, Y, distanc, cmap=cm.coolwarm, linewidth=0, antialiased=True)   
ax.view_init(30, 45)
```

```python
dist_graf = np.min(distanc)
xx,yy = np.where(distanc == dist_graf)
dis = distanc[xx,yy]
print("Distancia calculada:",dis)
```

Ambas distancias son aproximadamente iguales (2.53554728 y 2.5354627641855494), probando así que el método de mínimos cuadrados nos da la mínima distancia posible.
