Usando Newton-Raphson en 3.50

```python
def F(x):
  f1 = np.log(x[0]**2+x[1]**2) - np.sin(x[0]*x[1]) - np.log(2) - np.log(np.pi)
  f2 = np.exp(x[0]-x[1]) + np.cos(x[0]*x[1])
  return [f1,f2]

def dF(x):
  df1_x = (2*x[0]/x[0]**2+x[1]**2) - x[1]*np.cos(x[0]*x[1])
  df1_y = (2*x[0]/x[0]**2+x[1]**2) - x[1]*np.cos(x[0]*x[1])
  df2_x = np.exp(x[0]-x[1]) - x[1]*np.sin(x[0]*x[1])
  df2_y = -np.exp(x[0]-x[1]) - x[0]*np.sin(x[0]*x[1])
  return [[df1_x,df1_y],[df2_x,df2_y]]
```

```python
N = 100
x = [2,2]

for i in range(N):
  xold = x
  Jinv = np.linalg.inv(dF(x))
  x = x-np.dot(Jinv,F(x))
  e = np.linalg.norm(x-xold)
  print(i,x,e)
  if e<1e-10:
    break
```
