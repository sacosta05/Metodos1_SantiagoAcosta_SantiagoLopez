```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1,1.1,400)
h = 0.01

def derivada_progresiva(x,f,h):
    return (f(x+h)-f(x))/h

def derivada_central(f, x, h):
    return (f(x+h) - f(x-h)) / (2*h)

def f(x):
    return np.sqrt(np.tan(x))

def derivada_real(x):
    return (1/(np.cos(x)**2))/(2*np.sqrt(np.tan(x)))
```

```python
plt.plot(x,derivada_central(f,x,h),c='darkred', linewidth=2)
plt.title("Funci√≥n f '(x)", size=20, c='darkred')
plt.show()
```

```python
plt.plot(x,derivada_real(x),c='darkred', linewidth=2)
plt.title("Funci√≥n f '(x)", size=20, c='darkred')
plt.show()
```

```python
plt.plot(x,np.abs(derivada_real(x)-derivada_central(f,x,h)), c='darkgreen', linewidth=2)
plt.title("Estimaci√≥n del error en cada punto nodal respecto a la derivada central.", size=18, c='darkgreen')
plt.show()
```
